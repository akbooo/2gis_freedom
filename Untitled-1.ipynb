```python
import requests
import json
import time


API_KEY = "c7f1a769-c8a5-4636-b14d-d8c987808a12"
BRANCH_ID = "70000001049041638"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36",
    "Referer": "https://2gis.kz/",
    "Origin": "https://2gis.kz",
}

def get_reviews_list(branch_id, limit=50):
    all_reviews = []
    offset_date = None

    while True:
        url = f"https://public-api.reviews.2gis.com/2.0/branches/{branch_id}/reviews"
        params = {
            "key": API_KEY,
            "fields": "meta.total_count",
            "locale": "ru_KZ",
            "limit": limit,
            "sort_by": "date_created",  # ← исправлено
        }
        if offset_date:
            params["offset_date"] = offset_date

        r = requests.get(url, params=params, headers=headers, timeout=10)
        print(f"Статус: {r.status_code} | Ответ: {r.text[:300]}")

        if r.status_code != 200:
            break

        data = r.json()
        reviews = data.get("reviews", [])
        if not reviews:
            break

        all_reviews.extend(reviews)
        total = data.get("meta", {}).get("total_count", "?")
        print(f"Отзывов собрано: {len(all_reviews)} / {total}")

        if len(reviews) < limit:
            break

        # Пагинация по дате последнего отзыва
        offset_date = reviews[-1].get("date_created") or reviews[-1].get("created")
        time.sleep(1)

    return all_reviews

reviews = get_reviews_list(BRANCH_ID)

with open("freedom_bank_ast.json", "w", encoding="utf-8") as f:
    json.dump(reviews, f, ensure_ascii=False, indent=2)

print(f"\n✅ Итого: {len(reviews)} отзывов")
```

```python
def get_key(name):
    parts = name.split()

    if len(parts) >= 2:
        # фамилия = первый токен (очень типично для RU/KZ датасетов)
        return parts[0]
    return name
from rapidfuzz import process, fuzz

def fuzzy_normalize(names, threshold=88):
    canonical = []

    mapping = {}

    for name in names:
        if not canonical:
            canonical.append(name)
            mapping[name] = name
            continue

        match, score, _ = process.extractOne(name, canonical, scorer=fuzz.WRatio)

        if score >= threshold:
            mapping[name] = match
        else:
            canonical.append(name)
            mapping[name] = name

    return mapping
```
